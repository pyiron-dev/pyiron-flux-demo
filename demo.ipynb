```python
import subprocess
```

```python
script_str = """\
import flux.job
from pyiron_atomistics import Project

pr = Project('test')
job = pr.create.job.Lammps("lmp")
job.executable.version = "2020.03.03_flux"
job.structure = pr.create.structure.ase.bulk("Al", cubic=True)
job.flux_executor = flux.job.FluxExecutor()
job.server.cores = 2
fs = job.run()

print(fs.done())
print(fs.result())
print(fs.done())
"""
```

```python
with open("demo.py", "w") as f:
    f.writelines(script_str)
```

```python
subprocess.check_output(["flux", "start", "python", "demo.py"], universal_newlines=True).split("\n")
```
